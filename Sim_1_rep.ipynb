## Performance

### Large $\sigma$

```python
alpha=0
kappa_buy=27
kappa_sell=27
lambda_buy=2
lambda_sell=2
sigma=0.5
q_upper=15
q_lower=-15
Time=300
dt=0.005
theta=0.001
phi=0
phi_alpha=10
phi_lambda=20
phi_kappa=20

agent = RobustAgent(
    alpha=alpha,
    kappa_buy=kappa_buy,
    kappa_sell=kappa_sell,
    lambda_buy=lambda_buy,
    lambda_sell=lambda_sell,
    sigma=sigma,
    q_upper=q_upper,
    q_lower=q_lower,
    Time=Time,
    dt=dt,
    theta=theta,
    phi=phi,
    phi_alpha=phi_alpha,
    phi_kappa=phi_kappa,
    phi_lambda=phi_lambda,
)
```

```python
MC_samples = 1000

MC_reward_robust = []
MC_reward_plain = []

for _ in range(MC_samples):
    agent.complex_env_sim(control_uncertainty=True)
    MC_reward_robust.append(agent.objective)
    agent.reset()
    agent.complex_env_sim(control_uncertainty=False)
    MC_reward_plain.append(agent.objective)
    agent.reset()

MC_reward_robust = np.array(MC_reward_robust)
MC_reward_plain = np.array(MC_reward_plain)
```

```python
plt.plot(agent.time_grid, np.mean(MC_reward_robust, axis=0), label="Robust")
plt.plot(agent.time_grid, np.mean(MC_reward_plain, axis=0), label="Plain")

plt.xlabel("Time (secs)")
plt.ylabel("Reward")
plt.legend()
plt.show()
```

### Sharp ration

```python
alpha=0
kappa_buy=27
kappa_sell=27
lambda_buy=2
lambda_sell=2
sigma=0.01
q_upper=8
q_lower=-8
Time=300
dt=0.005
theta=0.001
phi=0
phi_alpha=0
phi_lambda=np.arange(0, 25, 5)

expectation_robust = []
expectation_plain = []
standard_deviation_robust = []
standard_deviation_plain = []

for phi_lambda_ in phi_lambda:
    agent = RobustAgent(
        alpha=alpha,
        kappa_buy=kappa_buy,
        kappa_sell=kappa_sell,
        lambda_buy=lambda_buy,
        lambda_sell=lambda_sell,
        sigma=sigma,
        q_upper=q_upper,
        q_lower=q_lower,
        Time=Time,
        dt=dt,
        theta=theta,
        phi=phi,
        phi_alpha=phi_alpha,
        phi_kappa=phi_lambda_,
        phi_lambda=phi_lambda_,
    )

    MC_samples = 100
    MC_reward_robust = []
    MC_reward_plain = []
    
    for _ in range(MC_samples):
        agent.complex_env_sim(control_uncertainty=True)
        MC_reward_robust.append(agent.objective[-1])
        agent.reset()
        agent.complex_env_sim(control_uncertainty=False)
        MC_reward_plain.append(agent.objective[-1])
        agent.reset()

    expectation_robust.append(np.mean(np.array(MC_reward_robust)))
    expectation_plain.append(np.mean(np.array(MC_reward_plain)))
    standard_deviation_robust.append(np.std(np.array(MC_reward_robust)))
    standard_deviation_plain.append(np.std(np.array(MC_reward_plain)))

```

```python
plt.plot(standard_deviation_robust, expectation_robust, 'x-',label="Robust")
plt.plot(standard_deviation_plain, expectation_plain, 'x-',label="Plain")

plt.xlabel("Standard Deviation")
plt.ylabel("Expectation")
plt.legend()
plt.show()
```
